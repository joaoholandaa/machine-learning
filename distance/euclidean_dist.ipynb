```python
import numpy as np

x1 = np.array([1,2])
x2 = np.array([3,1])
```

```python
def emat(p, q):
  import math
  return np.sqrt(math.pow((p[0] - q[0]), 2) + math.pow((p[1] - q[1]), 2))

emat(x1, x2)
```

```python
def sqrt_sum(p, q):
  return np.sqrt(np.sum((p-q)**2, axis=0))

sqrt_sum(x1, x2)
```

```python
def bymatmul(p,q):
  return np.sqrt((p-q) @ (p-q))

bymatmul(x1, x2)
```

```python
def bymatmul2(p,q):
  return np.sqrt(np.matmul(np.transpose(p-q),p-q))

bymatmul2(x1, x2)
```

```python
def linalg_norm(p,q):
  return np.linalg.norm(p-q)

linalg_norm(x1, x2)
```

```python
def edot(p,q):
  return np.sqrt(np.dot(p-q, p-q))

edot(x1, x2)
```

```python
def scipy_distance(p,q):
  from scipy.spatial import distance
  return distance.euclidean(p, q)

scipy_distance(x1, x2)
```
