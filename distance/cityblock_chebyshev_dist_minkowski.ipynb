**CityBlock**

```python
import numpy as np

x1 = np.array([1,2])
x2 = np.array([3,1])
```

```python
def emat(p,q):
  return pow(pow(abs(p[0] - q[0]), 1) + pow(abs(p[1] - q[1]), 1), 1/1)

emat(x1, x2)
```

```python
def sqrt_sum(p,q):
    return pow(np.sum(abs(p-q)**1, axis=0), 1/1)

sqrt_sum(x1,x2)
```

```python
def edot(p,q):
    return np.sum(abs(p-q))

edot (x1,x2)
```

```python
def scipy_distance(p,q):
    from scipy.spatial import distance
    return distance.cityblock(p,q)

scipy_distance(x1,x2)
```

**Chebyshev**

```python
import numpy as np

x1 = np.array([1,2])
x2 = np.array([3,1]) 
```

```python
def sqrt_sum(p,q):
    return max(abs(p-q))

sqrt_sum(x1,x2)
```

```python
def scipy_distance(p,q):
    from scipy.spatial import distance
    return distance.chebyshev(p,q)

scipy_distance(x1,x2)
```
