```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as st
```

```python
def univariate_normal(x, mean, std):
    """pdf of the univariate normal distribution."""
    return ((1. / np.sqrt(2 * np.pi * std**2)) * np.exp(-(x - mean)**2 / (2 * std**2)))
```

```python
# Plot different Univariate Normals
x = np.linspace(-5, 5, num=150)
fig = plt.figure(figsize=(10, 5))
plt.plot(x, univariate_normal(x, mean=0, std=1),label="N(0,1)")
plt.xlabel('', fontsize=13)
plt.ylabel('density: ', fontsize=13)
plt.title('Univariate normal distributions')
plt.ylim([0, 1])
plt.xlim([-5, 5])
plt.legend(loc=1)

plt.show()
```

```python
# Plot different Univariate Normals
x = np.linspace(-5, 5, num=150)
fig = plt.figure(figsize=(10, 7))
plt.plot(x, univariate_normal(x, mean=0, std=2),label="N(0, 2")
plt.plot(x, univariate_normal(x, mean=0, std=0.5),label="N(0, 0.5)")
plt.plot(x, univariate_normal(x, mean=2, std=2),label="N(2, 2)")
plt.plot(x, univariate_normal(x, mean=2, std=0.4),label="N(2, 0.4)")
plt.xlabel('x', fontsize=13)
plt.ylabel('density: p(x)', fontsize=13)
plt.title('Univariate normal distributions')
plt.ylim([0, 1])
plt.xlim([-5, 5])
plt.legend(loc=1)

plt.show()
```

```python
# Plot different Univariate Normals
x = np.linspace(-5, 5, num=150)
fig = plt.figure(figsize=(10, 7))
plt.plot(x, st.norm.pdf(x, loc=0, scale=1),label="N(0, 1)")
plt.plot(x, st.norm.pdf(x, loc=2, scale=2),label="N(2, 3)")
plt.plot(x, st.norm.pdf(x, loc=0, scale=0.5),label="N(0, 0.2)")
plt.xlabel('x', fontsize=13)
plt.ylabel('density: p(x)', fontsize=13)
plt.title('Univariate normal distributions')
plt.ylim([0, 1])
plt.xlim([-5, 5])
plt.legend(loc=1)
plt.show()
```
